```python
import numpy as np
import matplotlib.pyplot as plt
from itertools import combinations
from matplotlib.colors import LinearSegmentedColormap
work_dir = '/Users/oskarsvensson/Science/FILES/ARTICLES/Hst1_self/Hst1_self_asso/analysis/convergence/'

cm = 1/2.54
font = 8
font_label = 10
plt.rcParams['font.family'] = 'Helvetica'

colors = ['darkblue',  'aquamarine', 'darkorange']
cmap = LinearSegmentedColormap.from_list('mycmap', colors)
amount = 100
color_list = cmap(np.linspace(0, 1, amount))
```

```python
fig, ax = plt.subplots(ncols=3, nrows=2, figsize=((24*cm), (12*cm)), layout='constrained')
pairs = list(combinations(range(4), 2))
axis_list = [(0, 0), (0, 1), (0, 2), (1, 0), (1, 1), (1, 2)]

for (entry, nr) in zip(axis_list, range(1, 7, 1)):
    color_pick = 0
    for pair in pairs:
        t, dist = np.genfromtxt((work_dir + 't' + str(nr) + '_' + str(pair[0] + 1) + '_' + str(pair[1] + 1) + '.xvg'), 
                                usecols=(0, 1), skip_header=24, unpack=True)
        ax[entry[0], entry[1]].plot((t*(10**-6)), dist, color=color_list[color_pick], label=(str(pair[0] + 1) + '-' + str(pair[1] + 1)))
        ax[entry[0], entry[1]].tick_params(labelsize=font)
        if nr == 1:
            ax[entry[0], entry[1]].legend(fontsize=font)
        color_pick += 19

ax[0, 0].set_ylabel('Pair Distance [nm]', fontsize=font_label)
ax[1, 0].set_ylabel('Pair Distance [nm]', fontsize=font_label)
ax[1, 0].set_xlabel('Time [μs]', fontsize=font_label)
ax[1, 1].set_xlabel('Time [μs]', fontsize=font_label)
ax[1, 2].set_xlabel('Time [μs]', fontsize=font_label)

ax[0, 0].set_title('A', fontsize=font)
ax[0, 1].set_title('B', fontsize=font)
ax[0, 2].set_title('C', fontsize=font)
ax[1, 0].set_title('D', fontsize=font)
ax[1, 1].set_title('E', fontsize=font)
ax[1, 2].set_title('F', fontsize=font)
plt.savefig('/Users/oskarsvensson/Desktop/Current/Hst1_asso/New_new_figures/FA_pairdist_convergence.pdf', dpi=1000)
```
